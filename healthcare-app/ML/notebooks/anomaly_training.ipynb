# Import Modules

```python
import numpy as np
from sklearn.model_selection import train_test_split
import joblib
import os

from src.preprocessing import generate_synthetic_data, scale_data
from src.autoencoder_model import build_autoencoder
from src.isolation_model import train_isolation_forest
```

# Generate Synthetic Data

```python
data = generate_synthetic_data(5000)
scaled_data = scale_data(data)
```

# Train/Test Split

```python
X_train, X_test = train_test_split(
    scaled_data,
    test_size=0.2,
    random_state=42
)
```

# Train Isolation Forest

```python
iso_model = train_isolation_forest(X_train)
 
```

# Train Autoencoder

```python
autoencoder = build_autoencoder(X_train.shape[1])

autoencoder.fit(
    X_train,
    X_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)


```

# Compute Threshold & Save

```python
reconstructions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - reconstructions, 2), axis=1)

threshold = np.percentile(mse, 95)


print("Training complete.")
print("Threshold:", threshold)

```
